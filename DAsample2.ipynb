```python
def binary_search(arr, target):
    """
    Implements Binary Search and counts the number of probes.
    """
    low = 0
    high = len(arr) - 1
    probes = 0

    while low <= high:
        probes += 1
        mid = (low + high) // 2

        if arr[mid] == target:
            return mid, probes
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1

    return -1, probes


def interpolation_search(arr, target):
    """
    Implements Interpolation Search and counts the number of probes.
    """
    low = 0
    high = len(arr) - 1
    probes = 0

    # The target must be within the bounds of the array
    while low <= high and arr[low] <= target <= arr[high]:
        probes += 1

        # Guard against division by zero if all remaining elements are identical
        if arr[high] == arr[low]:
            if arr[low] == target:
                return low, probes
            break

        # Estimate the position using the interpolation formula
        # pos = low + [ (target - arr[low]) * (high - low) / (arr[high] - arr[low]) ]
        pos = low + int(((float(high - low) / (arr[high] - arr[low])) * (target - arr[low])))

        if arr[pos] == target:
            return pos, probes
        elif arr[pos] < target:
            low = pos + 1
        else:
            high = pos - 1

    return -1, probes


# --- Execution and Comparison ---
if __name__ == "__main__":
    import random

    # 1. Generate a sorted array of unique student roll numbers between 1 and 10000
    # Let's say we have 5000 students distributed across this range
    array_size = 5000
    student_rolls = sorted(random.sample(range(1, 10001), array_size))

    # 2. Pick a target that is guaranteed to be in the array for a fair test
    target_index = random.randint(0, array_size - 1)
    target_roll = student_rolls[target_index]

    print(f"Array Size (Total Students): {array_size}")
    print(f"Searching for Roll Number: {target_roll}\n")
    print("-" * 40)

    # 3. Run Binary Search
    bs_idx, bs_probes = binary_search(student_rolls, target_roll)
    print(f"Binary Search Result:")
    print(f"  Found at index: {bs_idx}")
    print(f"  Probes required: {bs_probes}")

    print("-" * 40)

    # 4. Run Interpolation Search
    is_idx, is_probes = interpolation_search(student_rolls, target_roll)
    print(f"Interpolation Search Result:")
    print(f"  Found at index: {is_idx}")
    print(f"  Probes required: {is_probes}")
    print("-" * 40)

```
